```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import numpy as np
from matplotlib import font_manager
```

```python
%matplotlib inline
```

```python
fontP = font_manager.FontProperties()
fontP.set_family('Heiti TC')
```

```python
df = pd.read_csv("all_analyzed.csv")
df.head()
```

# Size

```python
ratio = df.groupby('member_casual').size() / len(df) * 100
ratio
```

```python
fig_ratio, ax_ratio = plt.subplots(figsize=(6, 5))
ratio.plot.bar(ax=ax_ratio, color=['#294979', '#F19D38'])

# set title
ax_ratio.set_title('會員與非會員佔比', fontsize=22, fontweight='bold', x=0.19, pad=50, fontproperties=fontP)

# set x axis
ax_ratio.set_xticklabels(["非會員","會員"], rotation=0, fontsize='large', fontproperties=fontP) 
ax_ratio.set_xlabel('')

# hide spines
ax_ratio.spines['top'].set_visible(False)
ax_ratio.spines['right'].set_visible(False)
ax_ratio.spines['left'].set_visible(False)

# add percentage above bars
for i, v in enumerate(ratio):
    if i % 2 == 1:
        ax_ratio.text(i-0.09, v+3, str(format(v, '.1f'))+'%', color='#F19D38', fontweight='bold', fontsize='large')
    else:
        ax_ratio.text(i-0.09, v+2.5, str(format(v, '.1f'))+'%', color='#294979', fontweight='bold', fontsize='large')
```

## Conclusion
Member has a bigger portion

# Distance

```python
dist = df.groupby('member_casual').agg({'distance': ['mean', 'median', 'max', 'min']})
dist
```

```python
dist = dist.reset_index()
```

```python
dist_member = df[df['member_casual']=='member']['distance'].dropna()
dist_casual = df[df['member_casual']=='casual']['distance'].dropna()
```

```python
fig_dist, ax_dist = plt.subplots()
boxplot_dist = ax_dist.boxplot([dist_member, dist_casual], showfliers=False, patch_artist=True) # no fliers

ax_dist.set_title('會員與非會員的距離分佈箱形圖', fontsize=22, fontweight='bold', x=0.37, pad=40, fontproperties=fontP)

# fill box color
colors = ['#294979','#F19D38']
for patch, color in zip(boxplot_dist['boxes'], colors):
        patch.set_facecolor(color)

# add median line color
for median in boxplot_dist['medians']:
    median.set_color('black')

# set x axis
ax_dist.set_xticklabels(["非會員","會員"], rotation=0, fontsize='large', fontproperties=fontP) 
ax_dist.set_xlabel('')

# set y axis label location
ax_dist.set_ylabel('公里', rotation="horizontal", fontproperties=fontP)
ax_dist.yaxis.set_label_coords(-0.02, 0.98)

# hide spines
ax_dist.spines['top'].set_visible(False)
ax_dist.spines['right'].set_visible(False)
ax_dist.spines['left'].set_visible(False)
ax_dist.spines['bottom'].set_visible(False)
```

## Conclusion
No big difference between casual and member

# Length of ride

```python
length = df.groupby('member_casual').agg({'ride_length': ['mean', 'median']})
length
```

```python
length = length.reset_index()
```

```python
member_casual = length['member_casual']
mean_length = [length['ride_length']['mean'][0], length['ride_length']['mean'][1]]
median_length = [length['ride_length']['median'][0], length['ride_length']['median'][1]]

x = np.arange(len(member_casual))

fig_length, ax_length = plt.subplots()
ax_length.bar(x-0.15, mean_length, 0.35, color='#294979')
ax_length.bar(x+0.2, median_length, 0.35, color='#F19D38')

ax_length.set_title('會員與非會員的平均和中位數時間', fontsize=22, fontweight='bold', x=0.38, pad=50, fontproperties=fontP)

ax_length.get_xaxis().set_visible(False)
ax_length.set_ylabel('分鐘', rotation="horizontal", fontproperties=fontP)
ax_length.yaxis.set_label_coords(-0.04, 0.94)

ax_length.spines['top'].set_visible(False)
ax_length.spines['right'].set_visible(False)
ax_length.spines['left'].set_visible(False)

ax_length.text(-0.27, 29, '非會員',  fontsize='x-large', color='#294979', fontproperties=fontP)
ax_length.text(0.73, 13.4, '非會員',  fontsize='x-large', color='#294979', fontproperties=fontP)

ax_length.text(0.12, 13.3, '會員',  fontsize='x-large', color='#F19D38', fontproperties=fontP)
ax_length.text(1.11, 9.55, '會員',  fontsize='x-large', color='#F19D38', fontproperties=fontP)

ax_length.text(-0.14, -2, '平均時間',  fontsize='x-large', fontproperties=fontP)
ax_length.text(0.84, -2, '中位數時間',  fontsize='x-large', fontproperties=fontP)
```

## Conclusion
Casuals spend more time than members

# difference across year

```python
df['started_at'] = pd.to_datetime(df['started_at'])
df['month'] = df['started_at'].dt.strftime('%Y-%m')
```

```python
df.groupby(['member_casual', 'month']).size() 
```

```python
month = df.groupby(['member_casual', 'month']).size() / df.groupby(['month']).size() * 100
month
```

```python
ax_month = month.unstack('member_casual').plot(color=['lightgrey', (0.94, 0.61, 0.22)], marker='.', markevery=[0, 6], markersize=10)

ax_month.set_title('會員與非會員一年中每月所佔比例', fontweight='bold', fontsize=22,  x=0.393, pad=40,  fontproperties=fontP)
ax_month.set_ylim(0, 85)
ax_month.set_xlabel('')
# hide all spines
ax_month.spines['top'].set_visible(False)
ax_month.spines['right'].set_visible(False)
ax_month.spines['left'].set_visible(False)
# ax_rush_hour.spines['bottom'].set_visible(False)

# remove legend
ax_month.get_legend().remove()

# set the essential on the plot
for i, v in enumerate(month):
    if i % 6 == 0:
        if i > 6:
            ax_month.text(i-12.2, v+3, str(int(v))+'%', color=(0.94, 0.61, 0.22), fontweight='bold', fontsize='large')
        else:
            ax_month.text(i-0.2, v-6, str(int(v))+'%', color='lightgrey', fontweight='bold', fontsize='large')

ax_month.text(11.2, 58, '會員', color=(0.94, 0.61, 0.22), alpha=0.7, fontweight='bold', fontsize='x-large',  fontproperties=fontP)
ax_month.text(11.2, 41, '非會員', color='lightgrey', fontweight='bold', fontsize='x-large',  fontproperties=fontP)
```

## Conclusion
The percentage of casual sharply decreases when it comes to winter (49%->21%).

# Work day

```python
df['hour'] = df['started_at'].dt.strftime('%H')
df['hour']
```

```python
work_day_hour = df[df['work_day']==1].pivot_table(index='hour', columns='member_casual', aggfunc='size')
not_work_day_hour = df[df['work_day']==0].pivot_table(index='hour', columns='member_casual', aggfunc='size')
```

```python
fig_hour, ax_hour = plt.subplots(1, 2, figsize=(10, 5))

### WORK DAY ###
ax_work_day_hour = work_day_hour.plot(ax=ax_hour[0], color=['#294979', '#F19D38'])

# title setting
ax_work_day_hour.set_title('上班日', fontsize=14, pad=20, fontweight='bold', fontproperties=fontP)

# hide all spines
ax_work_day_hour.spines['top'].set_visible(False)
ax_work_day_hour.spines['right'].set_visible(False)
ax_work_day_hour.spines['left'].set_visible(False)

ax_work_day_hour.set_xlabel('小時', fontsize=10, fontproperties=fontP)
ax_work_day_hour.set_ylim(0, 350000)
ax_work_day_hour.set_ylabel('(人數)', rotation="horizontal", fontproperties=fontP)
ax_work_day_hour.yaxis.set_label_coords(-0.06, 1.1)

ax_work_day_hour.text(0.6, 100000, '會員', color='#F19D38',  fontweight='bold', fontsize='x-large', fontproperties=fontP)
ax_work_day_hour.text(14.5, 50000, '非會員', color='#294979', fontweight='bold', fontsize='x-large', fontproperties=fontP)

# remove legend
ax_work_day_hour.get_legend().remove()

ax_work_day_hour.arrow(8, 218000, 0, -8000, head_width=0.5, head_length=8000, width=0.2, color='#F19D38')
ax_work_day_hour.arrow(17, 328000, 0, -8000, head_width=0.5, head_length=8000, width=0.2, color='#F19D38')
ax_work_day_hour.text(6.9, 222000, '8:00', color='#F19D38', fontweight='bold', fontsize=10)
ax_work_day_hour.text(15.7, 332000, '17:00', color='#F19D38', fontweight='bold', fontsize=10)

### NOT WORK DAY ###
ax_not_work_day_hour = not_work_day_hour.plot(ax=ax_hour[1], color=['#294979', '#F19D38'])

# title setting
ax_not_work_day_hour.set_title('非上班日',  fontsize=14, pad=20, fontweight='bold', fontproperties=fontP)

# hide all spines
ax_not_work_day_hour.spines['top'].set_visible(False)
ax_not_work_day_hour.spines['right'].set_visible(False)
ax_not_work_day_hour.spines['left'].set_visible(False)

ax_not_work_day_hour.set_xlabel('小時', fontsize=10, fontproperties=fontP)
ax_not_work_day_hour.set_ylim(0, 350000)
ax_not_work_day_hour.get_yaxis().set_visible(False)

# remove legend
ax_not_work_day_hour.get_legend().remove()

ax_not_work_day_hour.text(5.6, 60000, '會員', color='#F19D38',  fontweight='bold', fontsize='x-large', fontproperties=fontP)
ax_not_work_day_hour.text(10.5, 30000, '非會員', color='#294979', fontweight='bold', fontsize='x-large', fontproperties=fontP)

fig_hour.suptitle("會員與非會員在上班日時段比較", fontsize=22, fontweight='bold', x=.227,y=1.09, fontproperties=fontP)
fig_hour.tight_layout()
```

```python
work_day = df.groupby(['work_day', 'member_casual']).size() / df.groupby('work_day').size() * 100
work_day
```

```python
ax_work_day = work_day.unstack('member_casual').plot(figsize=(7, 4),color=['lightgrey', (0.94, 0.61, 0.22)], linewidth=4, marker='.', markersize=16)

# title setting
ax_work_day.set_title('會員與非會員在上班日和非上班日比較', fontweight='bold', fontsize=22,  x=0.47, pad=50, fontproperties=fontP)

# set the range of x and y axis
ax_work_day.set_xlim(-0.2, 1.2)
ax_work_day.set_ylim(10, 70)

# hide x and y axis
ax_work_day.get_xaxis().set_visible(False)
ax_work_day.get_yaxis().set_visible(False)

# hide all spines
ax_work_day.spines['top'].set_visible(False)
ax_work_day.spines['right'].set_visible(False)
ax_work_day.spines['left'].set_visible(False)

# remove legend
ax_work_day.get_legend().remove()

# set the essential on the plot
for i, v in enumerate(work_day):
    if i % 2 == 1:
        ax_work_day.text(int(i/2)-0.03, v+3, str(int(v))+'%', color=(0.94, 0.61, 0.22), alpha=0.7, fontweight='bold', fontsize='large')
    else:
        ax_work_day.text(int(i/2)-0.03, v-6, str(int(v))+'%', color='lightgrey', fontweight='bold', fontsize='large')

ax_work_day.text(-0.155, 51, '會員', color=(0.94, 0.61, 0.22), alpha=0.7, fontweight='bold', fontsize='x-large', fontproperties=fontP)
ax_work_day.text(-0.2, 47, '非會員', color='lightgrey', fontweight='bold', fontsize='x-large', fontproperties=fontP)

ax_work_day.text(-0.09, 6, '非上班日',  fontsize='x-large', fontproperties=fontP)
ax_work_day.text(0.93, 6, '上班日',  fontsize='x-large', fontproperties=fontP)
```

## Conclusion
On non-work day, the percentages between members and casuals are almost the same, but when it comes to work day, the percentage of members increases.
